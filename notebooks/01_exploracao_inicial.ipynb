A análise da criminalidade é essencial para entender padrões e auxiliar na tomada de decisão em segurança pública.

Este projeto busca explorar dados de ocorrências na cidade de Teresina, com o objetivo de identificar padrões temporais e geográficos, respondendo perguntas como:

- Quais crimes são mais recorrentes?
- Quando eles acontecem com maior frequência?
- Onde estão mais concentrados?

## 📊 Objetivo

Desenvolver uma análise exploratória de dados e construir visualizações que permitam identificar tendências e gerar insights relevantes.

## 📌 Hipóteses

- Crimes ocorrem com maior frequência no período noturno
- Finais de semana apresentam maior número de ocorrências
- Alguns bairros concentram tipos específicos de crime

## 🟣 1. Importação de Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configuração visual
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## 🟣 2. Carregamento dos Dados

```python
caminho = '../data/raw/'
arquivos = os.listdir(caminho)

dfs = []

for arquivo in arquivos:
    if arquivo.endswith('.xlsx'):
        df_temp = pd.read_excel(caminho + arquivo)
        df_temp['origem_arquivo'] = arquivo
        dfs.append(df_temp)

df = pd.concat(dfs, ignore_index=True)

df.head()
```

## 🚨 Problemas a serem resolvidos

O objetivo deste projeto é consolidar diferentes bases de dados de ocorrências criminais em um único dataset estruturado, possibilitando análises consistentes e confiáveis.

Entretanto, devido à integração de múltiplas fontes com estruturas distintas, foram identificados alguns desafios que precisam ser tratados previamente:

- Inconsistência na nomenclatura e tipagem das colunas  
- Existência de colunas com a mesma finalidade, porém com nomes diferentes  
- Presença de valores nulos decorrentes da falta de padronização entre as bases  

Dessa forma, torna-se necessário realizar um processo de padronização e limpeza dos dados, garantindo a uniformidade das colunas e a qualidade das informações antes da etapa de análise.
