# 📊 Análise de Criminalidade em Teresina

## 🎯 Definição do Problema
Este projeto tem como objetivo analisar os padrões de criminalidade na cidade de Teresina, buscando identificar tendências relacionadas ao tipo de crime, localização, data e horário das ocorrências.

## ❓ Perguntas de Negócio
- Quais são os crimes mais frequentes?
- Em quais horários ocorrem com maior incidência?
- Quais bairros concentram mais ocorrências?
- Existe variação por dia da semana ou período?

## 🧠 Objetivo da Análise
Gerar insights que ajudem a compreender o comportamento da criminalidade e apoiar possíveis tomadas de decisão baseadas em dados.
