# 📊 Criação do Dataset Base

## 🎯 Objetivo

Realizar o tratamento e a padronização das bases de dados, unificando-as em um único dataset estruturado e pronto para análise.

## 🛠️ Etapas

Inicialmente, o **Banco de Dados 1** será dividido em dois conjuntos distintos, separando os registros de **roubos de celulares** e **roubos de veículos**. Em seguida, serão removidas as colunas desnecessárias de todas as bases, garantindo maior consistência e organização dos dados.

Por fim, todas as bases tratadas serão integradas em um único dataset, estruturado com colunas previamente definidas e padronizadas.

## 🟣 1. Importando Bibliotecas

```python
import pandas as pd
import os
```

## 🟣 2. Lendo o arquivo de roubos de celulares e veículos

```python
df = pd.read_excel('../data/raw/1. Roubo de celular e veículos.xlsx')
```

Verificando os valores da coluna:

```python
print(df["Unnamed: 6"].unique())
```

```python
df["Unnamed: 6"] = df["Unnamed: 6"].str.lower().str.strip()
```

## 🟣 3. Filtrar os dados (dividir em 2 datasets)

```python
df_celular = df[df["Unnamed: 6"] == "celulares"]
df_veiculo = df[df["Unnamed: 6"] == "veículo"]
```

Salvando os arquivos:

```python
os.makedirs('../data/processed', exist_ok=True)

df_celular.to_excel('../data/processed/roubo_celular.xlsx', index=False)
df_veiculo.to_excel('../data/processed/roubo_veiculo.xlsx', index=False)
```

## 🟣 4. Colocando todos os arquivos na pasta processed

```python
input_dir = '../data/raw'
output_dir = '../data/processed'

arquivos = os.listdir(input_dir)

arquivos = [arq for arq in arquivos if not arq.startswith('1.')]

for arquivo in arquivos:
    
    caminho_arquivo = os.path.join(input_dir, arquivo)
    
    df = pd.read_excel(caminho_arquivo)
    
    nome_base = arquivo.replace('.xlsx', '')
    nome_base = nome_base.lower()
    nome_base = nome_base.replace(' ', '_')
    nome_base = nome_base.replace('ç', 'c').replace('ã', 'a').replace('á', 'a')
    
    output_path = os.path.join(output_dir, f'{nome_base}.xlsx')
    df.to_excel(output_path, index=False)

    print(f'Arquivo processado: {arquivo} → {nome_base}.xlsx')
```

## 🟣 6. Juntando todos os datasets

```python
for i, df in enumerate([df0, df1, df2, df3, df4, df5, df6, df7, df8, df9]):
    print(f'\n--- df{i} ---')
    print(df.columns.tolist())
```

```python
df_ameaca = pd.read_excel('../data/processed/ameaca_filtrado.xlsx')
df_difamacao = pd.read_excel('../data/processed/difamacao_filtrado.xlsx')
df_embriaguez = pd.read_excel('../data/processed/embriaguez_ao_volante_filtrado.xlsx')
df_estelionato = pd.read_excel('../data/processed/estelionato_filtrado.xlsx')
df_estupro = pd.read_excel('../data/processed/estupro_filtrado.xlsx')
df_furto = pd.read_excel('../data/processed/furto_filtrado.xlsx')
df_lesao = pd.read_excel('../data/processed/lesao_corporal_filtrado.xlsx')

colunas_padrao = ['municipio_fato', 'logradouro_fato', 'bairro_final', 'dt_fato', 'hora', 'grupo_natureza']

df_roubo_celular = pd.read_excel('../data/processed/roubo_celular_filtrado.xlsx', header=None, skiprows=1)
df_roubo_celular.columns = colunas_padrao

df_roubo_veiculo = pd.read_excel('../data/processed/roubo_veiculo_filtrado.xlsx', header=None, skiprows=1)
df_roubo_veiculo.columns = colunas_padrao

df_violencia = pd.read_excel('../data/processed/violencia_domestica_filtrado.xlsx', header=2)
df_violencia = df_violencia.rename(columns={
    'Município Fato':    'municipio_fato',
    'Logradouro Fato':   'logradouro_fato',
    'Bairro Fato':       'bairro_final',
    'Data Início Fato':  'dt_fato',
    'Hora Início Fato':  'hora',
    'Grupo Natureza':    'grupo_natureza',
})

df_difamacao = df_difamacao.rename(columns={'Personalizar': 'hora'})

df_ameaca = df_ameaca.rename(columns={'HORA': 'hora'})

df_ameaca['crime']          = 'ameaca'
df_difamacao['crime']       = 'difamacao'
df_embriaguez['crime']      = 'embriaguez_ao_volante'
df_estelionato['crime']     = 'estelionato'
df_estupro['crime']         = 'estupro'
df_furto['crime']           = 'furto'
df_lesao['crime']           = 'lesao_corporal'
df_roubo_celular['crime']   = 'roubo_celular'
df_roubo_veiculo['crime']   = 'roubo_veiculo'
df_violencia['crime']       = 'violencia_domestica'

df_final = pd.concat([
    df_ameaca, df_difamacao, df_embriaguez, df_estelionato,
    df_estupro, df_furto, df_lesao, df_roubo_celular,
    df_roubo_veiculo, df_violencia
], ignore_index=True)

df_final = df_final[['crime', 'grupo_natureza', 'municipio_fato', 'bairro_final', 'logradouro_fato', 'dt_fato', 'hora']]

df_final.to_excel('../data/processed/banco_completo.xlsx', index=False)
print(df_final.shape)
print(df_final['crime'].value_counts())
```

## 🟣 7. Limpando o dataset final

```python
df_final.info()
df_final.isnull().sum()
df_final.nunique()
```

```python
def padronizar_hora(valor):
    if pd.isnull(valor):
        return None
    
    valor = str(valor).strip()
    
    # Caso já esteja completo
    if ':' in valor:
        try:
            return pd.to_datetime(valor).strftime('%H:%M:%S')
        except:
            return None
    
    # Caso seja só hora (ex: 20 → 20:00:00)
    if valor.isdigit():
        return f"{int(valor):02d}:00:00"
    
    return None

df_final['hora'] = df_final['hora'].apply(padronizar_hora)
```

```python
df_final['hora'] = pd.to_datetime(df_final['hora'], format='%H:%M:%S', errors='coerce')

df_final.to_excel('../data/processed/banco_completo.xlsx', index=False)
```
